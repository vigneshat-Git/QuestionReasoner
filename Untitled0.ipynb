```python
# Install libraries
!pip install transformers gradio torch

from transformers import pipeline, set_seed
import gradio as gr

# 🎯 Load GPT-2 model
generator = pipeline('text-generation', model='gpt2')
set_seed(42)

# 🧠 Story Generator Function
def local_generate_story(prompt, max_length=200):
    output = generator(prompt, max_length=max_length, num_return_sequences=1)
    return output[0]['generated_text']

# 🎨 Gradio UI
iface = gr.Interface(
    fn=local_generate_story,
    inputs=[
        gr.Textbox(label="Enter your story prompt"),
        gr.Slider(50, 300, value=150, label="Story Length")
    ],
    outputs="text",
    title="📚 Local Story Generator (GPT-2)",
    description="Offline text generation using Hugging Face GPT-2."
)

# 🚀 Launch UI
iface.launch()
```
